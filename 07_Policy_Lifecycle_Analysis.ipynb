## Module 7 — Policy Lifecycle Analysis

**Business Objective**
Understand when in the policy lifecycle claims are filed, and whether
timing is associated with claim outcomes or amounts. Findings inform
triage protocols, early warning systems, and underwriting review triggers.

**Analysis Scope**
Days-to-claim distribution · Claim timing buckets · Outcome by timing ·
Financial exposure by lifecycle stage · Early claim investigation

## 7.1 Days-to-Claim Distribution

```
claimed = get_claims_with_timing(df)
print(f"Policies with filed claims    : {len(claimed):,}")
print(f"\nDays-to-Claim Statistics:")
print(claimed['Days_to_Claim'].describe().round(1).to_string())
print(f"\nEarly claims (≤30 days)       : {(claimed['Days_to_Claim'] <= 30).sum():,}")
print(f"% of filed claims             : {(claimed['Days_to_Claim'] <= 30).mean():.1%}")
```

## 7.4 Early Claims Analysis

```
early = claimed[claimed['Days_to_Claim'] <= 30]
print(f"Early claims total           : {len(early):,}")
print(f"\nEarly Claim Status Distribution:")
print(early['ClaimStatus'].value_counts().to_string())
print(f"\nEarly Claim Amount Stats:")
print(early['ClaimAmount'].describe().round(2).to_string())
print(f"\nEarly Claims by Policy Type:")
print(early.groupby('PolicyType')['ClaimAmount'].agg(['count','mean']).round(2).to_string())
```

**Key Finding:** Early claims (filed within 30 days of inception) exhibit a
higher settlement rate than mid or late claims in this dataset. This runs
counter to the typical insurance industry pattern, where early claims are
treated with additional scrutiny due to potential pre-existing condition
or opportunistic claiming risk.

This finding should be validated against additional data features — particularly
claim reason and medical history — in a production environment before
drawing firm operational conclusions.
