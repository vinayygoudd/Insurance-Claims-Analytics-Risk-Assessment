# Insurance Claims Analytics & Risk Assessment
## Module 2 — Exploratory Data Analysis

**Business Objective**
Surface patterns in claim outcomes, premium pricing, and coverage utilisation
that inform underwriting strategy, product pricing decisions, and claims
processing priorities.

**Analysis Scope**
Claim status distribution · Policy type performance · Age and gender
segmentation · Premium distribution · Claim amount analysis · Coverage
utilisation · Executive dashboard


```
import sys
sys.path.insert(0, '..')
from src.eda import (
    load_cleaned_data, plot_claim_status_distribution,
    plot_claim_status_by_policy_type, plot_claim_status_by_age_group,
    plot_premium_distribution, plot_claim_amount_boxplot,
    plot_utilization_analysis, plot_executive_dashboard
)
from IPython.display import Image
import warnings
warnings.filterwarnings('ignore')

df = load_cleaned_data()
print(f"Dataset loaded: {len(df):,} records")
```

## 2.1 Claim Status Distribution

**Business Question:** What proportion of the portfolio is settled, rejected,
or pending — and does the current resolution rate indicate operational efficiency?


```
plot_claim_status_distribution(df)
Image('../visuals/claim_status_distribution.png')
```

**Finding:** Rejections account for 43.5% of all policies — the largest single category.
Settlement rate stands at 33.8%. The high rejection rate warrants investigation into
whether rejection criteria are consistently applied across product lines.


## 2.2 Claim Status by Policy Type

**Business Question:** Which product lines carry the highest rejection or pending rates,
and where should claims processing resource be prioritised?


```
plot_claim_status_by_policy_type(df)
Image('../visuals/policy_type_vs_claim_status.png')
```

**Finding:** Rejection rates are broadly consistent across product lines,
suggesting portfolio-level rejection patterns rather than product-specific issues.
Travel policies carry the highest claim volume by nature of the portfolio mix (41%).


## 2.3 Claim Outcomes by Age Group

**Business Question:** Are older customer cohorts experiencing systematically
different claim outcomes — indicating potential underwriting misalignment?


```
plot_claim_status_by_age_group(df)
Image('../visuals/age_group_analysis.png')
```

**Finding:** Claim status composition is broadly uniform across age groups,
suggesting that adjudication decisions are not age-driven at the portfolio level.


## 2.4 Premium Distribution Analysis

**Business Question:** Are premiums priced consistently within and across product
lines, or are there structural pricing anomalies requiring actuarial review?


```
plot_premium_distribution(df)
Image('../visuals/premium_distribution.png')
```

**Finding:** Premium distributions are near-uniform across all policy types,
suggesting undifferentiated pricing. Risk-adjusted pricing should reflect
the distinct risk profiles of each product line.


## 2.5 Claim Amount Distribution

**Business Question:** Do settled claims carry materially higher amounts than
rejected ones — and which product lines generate the greatest financial exposure?


```
plot_claim_amount_boxplot(df)
Image('../visuals/claim_amount_boxplot.png')
```

**Finding:** Claim amounts are distributed similarly across status groups,
indicating claim size alone does not drive rejection decisions. This is
analytically validated in Module 3 via ANOVA.


## 2.6 Coverage Utilisation

**Business Question:** Which segments are consuming the highest proportion of
their coverage — representing the greatest financial exposure for the insurer?


```
plot_utilization_analysis(df)
Image('../visuals/utilization_analysis.png')
```

**Finding:** Utilisation rates vary across product types and age groups.
High-utilisation segments represent concentrated exposure and should be
monitored for pricing adequacy and claims frequency trends.
