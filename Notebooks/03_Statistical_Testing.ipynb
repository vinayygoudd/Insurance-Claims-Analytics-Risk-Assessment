# Insurance Claims Analytics & Risk Assessment
## Module 3 — Statistical Testing

**Business Objective**
Move beyond visual patterns to statistically validated conclusions.
Each test is structured around a specific business question, with findings
framed in terms of decision-making impact rather than statistical theory.

**Significance Level:** α = 0.05 across all tests

**Tests Conducted**
| # | Test | Variables |
|---|---|---|
| 1 | Chi-Square | Policy Type vs Claim Status |
| 2 | Chi-Square | Gender vs Claim Status |
| 3 | One-Way ANOVA | Claim Amount by Claim Status |
| 4 | One-Way ANOVA | Premium Amount by Policy Type |
| 5 | Independent T-Test | Premium Amount by Gender |
| 6 | One-Way ANOVA | Coverage Ratio by Policy Type |


```
import sys
sys.path.insert(0, '..')
from src.statistical_testing import (
    load_cleaned_data,
    test_policy_type_vs_claim_status,
    test_gender_vs_claim_status,
    test_claim_amount_by_status,
    test_premium_by_policy_type,
    test_premium_by_gender,
    test_coverage_ratio_by_policy_type,
    compile_results_summary,
)
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

df = load_cleaned_data()
```

## 3.1 Chi-Square: Policy Type vs Claim Status

**Business Question:** Is claim outcome independent of policy type, or do
certain product lines carry structurally different resolution rates?

- **H₀:** Policy type and claim status are independent
- **H₁:** Policy type and claim status are associated


```
r1 = test_policy_type_vs_claim_status(df)
print(f"Test        : {r1['test']}")
print(f"Chi² stat   : {r1['statistic']}")
print(f"P-value     : {r1['p_value']}")
print(f"Decision    : {r1['decision']}")
print(f"\nBusiness Interpretation:")
print(r1['business_interpretation'])
print(f"\nContingency Table:")
print(r1['contingency_table'].to_string())
```

## 3.2 Chi-Square: Gender vs Claim Status

```
r2 = test_gender_vs_claim_status(df)
print(f"Test        : {r2['test']}")
print(f"Chi² stat   : {r2['statistic']}")
print(f"P-value     : {r2['p_value']}")
print(f"Decision    : {r2['decision']}")
print(f"\nBusiness Interpretation:")
print(r2['business_interpretation'])
```

## 3.3 ANOVA: Claim Amount by Claim Status

```
r3 = test_claim_amount_by_status(df)
print(f"Test        : {r3['test']}")
print(f"F-statistic : {r3['statistic']}")
print(f"P-value     : {r3['p_value']}")
print(f"Decision    : {r3['decision']}")
print(f"\nGroup Means:")
print(r3['group_means'].to_string())
print(f"\nBusiness Interpretation:")
print(r3['business_interpretation'])
```

## 3.4 ANOVA: Premium Amount by Policy Type

```
r4 = test_premium_by_policy_type(df)
print(f"Test        : {r4['test']}")
print(f"F-statistic : {r4['statistic']}")
print(f"P-value     : {r4['p_value']}")
print(f"Decision    : {r4['decision']}")
print(f"\nGroup Means:")
print(r4['group_means'].to_string())
print(f"\nBusiness Interpretation:")
print(r4['business_interpretation'])
```

## 3.5 T-Test: Premium Amount by Gender

```
r5 = test_premium_by_gender(df)
print(f"Test        : {r5['test']}")
print(f"T-statistic : {r5['statistic']}")
print(f"P-value     : {r5['p_value']}")
print(f"Male Mean   : ₹{r5['male_mean']:,.2f}")
print(f"Female Mean : ₹{r5['female_mean']:,.2f}")
print(f"Decision    : {r5['decision']}")
print(f"\nBusiness Interpretation:")
print(r5['business_interpretation'])
```

## 3.6 ANOVA: Coverage Ratio by Policy Type

```
r6 = test_coverage_ratio_by_policy_type(df)
print(f"Test        : {r6['test']}")
print(f"F-statistic : {r6['statistic']}")
print(f"P-value     : {r6['p_value']}")
print(f"Decision    : {r6['decision']}")
print(f"\nGroup Means:")
print(r6['group_means'].to_string())
print(f"\nBusiness Interpretation:")
print(r6['business_interpretation'])
```

## 3.7 Results Summary


```
summary = compile_results_summary([r1, r2, r3, r4, r5, r6])
pd.set_option('display.max_colwidth', 80)
summary[['Test', 'Statistic', 'P-Value', 'Decision']]
```

**Overall Finding:** None of the six tests reached statistical significance
at α = 0.05. This is a substantive business finding — it indicates that
claim outcomes, premium levels, and coverage ratios are distributed
uniformly across demographic and product-line dimensions in this portfolio.
This uniformity may reflect dataset construction or genuine portfolio balance,
and should be contextualised with additional data sources in production.
